```python
from random import randint
import timeit
# Бинарный поиск
def binary_search(a, value):
    mid = len(a) // 2
    low = 0
    high = len(a) - 1

    while a[mid] != value and low <= high:
        if value > a[mid]:
            low = mid + 1
        else:
            high = mid - 1
        mid = (low + high) // 2

    if low > high:
        return -1
    else:
        return mid

a = []
for i in range(10):
    a.append(randint(1, 50))
a.sort()
print(a)

# искомое число
value = int(input())
start_time = timeit.default_timer()
print(binary_search(a, value))
print(str((timeit.default_timer() - start_time)*1000) + "ms")

start_time = timeit.default_timer()
print(a.index(value))
print(str((timeit.default_timer() - start_time)*1000) + "ms")
```

```python
# Интерполяционный
from random import randint

def interpolation(val, array):
    low = 0
    high = len(array)-1
    while low <= high and val >= array[low] and val <= array[high]:
        m = low + (high-low)*(val-array[low])//(array[high]-array[low])
        if array[low] == array[high]:
            break
        if array[m] == val:
            return m
        elif array[m] < val:
            low = m+1
        else:
            high = m-1
    if array[low] == val:
        return low
    if array[high] == val:
        return high
    return -1

arr = []
for i in range(10):
    arr.append(randint(1, 50))
arr.sort()
print(arr)
x = int(input())
start_time = timeit.default_timer()
print(interpolation(x,arr))
print(str((timeit.default_timer() - start_time)*1000) + "ms")

start_time = timeit.default_timer()
print(arr.index(x))
print(str((timeit.default_timer() - start_time)*1000) + "ms")
```
