### Реализация стека

```python
class stack:
    def __init__(self):
        self.stacks=list()
    def __len__(self):
        return len(self.stacks)
        
    # добавление элемента в стек
    def add(self, el):
        self.stacks.append(el)
        
    #проверка стека на пустоту
    def testing_null(self):
        return len(self.stacks)==0
    
    #извлечение элемента из стека
    def pop(self):
        if self.testing_null():
            return None
        else:
            return self.stacks.pop()
    
    def peek(self):
        if self.testing_null():
            return None
        else:
            print(self.stacks[-1])
            return self.stacks[-1]
```

### Реализация Дека

```python
class deq:
    def __init__(self):
        self.deqs=list()
    

    #добавление в конец дека
    def add_left(self, el):
        self.deqs.insert(0, el)
        
    #добавление в начало дека
    def add_right(self, el):
        self.deqs.append(el)
        
    #проверка дека на пустоту
    def testing_null(self):
        return len(self.deqs)==0
    
    #извлечение элемента из начала дека
    def pop_right(self):
        if self.testing_null():
            return None
        else:
            return self.deqs.pop()
        
    def peek_left(self):
        if self.testing_null():
            return None
        else:
            return self.deqs[0]
        
        
    def peek_right(self):
        if self.testing_null():
            return None
        else:
            return self.deqs[-1]
        
    #извлечение элемента из конца дека
    def pop_left(self):
        if self.testing_null():
            return None
        else:
            return self.deqs.pop(0)
```

### Работа с файлами

```python
import copy
import codecs
#чтение и запись в файл

# Вывод из файла в стек
def add_with_stack(file):
    st=stack()
    try:
        f=open(file+'.txt','r')
        t=f.readlines()
        f.close()
        for i in range(len(t)):
            st.add(t[i])
        return st
    except:
        print("Error, не удалось найти файл")

        
# Вывод файла в переменную типа string
def add_with_str(file):
    rez=""
    try:
        f=codecs.open(file+'.txt', encoding='utf-8')
        t=f.readlines()
        f.close()
        for i in range(len(t)):
            rez=rez+t[i]
        return rez
    except:
        print("Error, не удалось найти файл")
        
# Вывод из файла в дек
# def add_with_deq(file):
#     st=deq()
#     try:
#         f=open(file+'.txt,'r')
#         t=f.readlines()
#         f.close()
#         for i in range(len(t)):
#             st.add_left(t[i])
#         return st
#     except:
#         print("Error, не удалось найти файл")

# Запись из дека в файл
def add_file(file, deq):
    try:
        f=open(file+'.txt',"w", encoding="utf-8")
        while not deq.testing_null():
            f.write(deq.pop_left())
        f.close()
    except:
        print("Ошибка")
        return None

def add_file_stack(file, stack):
    try:
        f=open(file+'.txt',"w", encoding="utf-8")
        while not stack.testing_null():
            f.write(stack.pop())
        f.close()
    except:
        print("Ошибка")
        return None

# Ввод из переменной в файл
def add_file_str(file, text):
    try:
        f=open(file+'.txt',"w", encoding="utf-8")
        f.write(text)
        f.close()
    except:
        print("Ошибка")
        return None
```

Отсортировать строки файла, содержащие названия книг, в алфавитном порядке с использованием двух деков.

```python
def task1(file_res, file_rezult):
    temp=add_with_str(file_res)
    st1=deq()
    for i in temp.split("\n"):
        st1.add_right(i)
    if st1==None:
        return
    st2=deq()
    while not st1.testing_null():
        el=st1.pop_left()
        while not st2.testing_null() and st2.peek_left()<el:
            st1.add_right(st2.pop_left())
        st2.add_left(el)
    
    add_file(file_rezult, copy.deepcopy(st2))
    return st2
            
r=task1("lab4/task1-input", "lab4/task1-output")
if r is not None:
    while not r.testing_null():
        print(r.pop_left())
```

### Задание 4
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс круглых скобок в тексте, используя стек.

```python
def task4(file):
    s=stack()
    st=add_with_str(file)
    if st is None:
        return
    for i in st:
        if i=='(':
            s.add(i)
        elif i==')':
            if s.testing_null():
                return False
            s.pop()
    return s.testing_null()
print(task4('lab4/task4-1'))
print(task4('lab4/task4-2'))
```
